```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
# load a csv file of tips data
tips = sns.load_dataset('tips') 

# Create a scatter plot with a regression line
sns.lmplot(x='total_bill', y='tip', data=tips)

# Add title
plt.title('Total Bill vs. Tip Amount')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a histogram with KDE for the 'total_bill' column
sns.histplot(tips['total_bill'], kde=True, color='blue')

# Add title and labels
plt.title('Distribution of Total Bill Amounts')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a line plot with a regression line
sns.lineplot(data=tips, x='size', y='tip')

# Add title
plt.title('Relationship Between Size and Tip')
plt.show()
```

```python
# Create a box plot to compare tips across different days
sns.boxplot(x='day', y='tip', data=tips)

# Add title
plt.title('Tip Amounts by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Tip Amount ($)')
plt.show()
```

```python
# Creating a violin plot to visualize the distribution of tips by day of the week
sns.violinplot(x='day', y='tip', data=tips, inner='quartile', palette='coolwarm')

# Add title and labels
plt.title('Distribution of Tip Amounts by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Tip Amount ($)')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Load the Iris dataset
iris_df = sns.load_dataset('iris')

# Display the first few rows of the dataset
iris_df.head()
```

```python
import matplotlib.pyplot as plt

# Create histograms for sepal length and petal length
plt.figure(figsize=(10, 5))

# Sepal Length
plt.subplot(1, 2, 1)
sns.histplot(iris_df['sepal_length'], kde=True, color='blue')
plt.title('Distribution of Sepal Length')

# Petal Length
plt.subplot(1, 2, 2)
sns.histplot(iris_df['petal_length'], kde=True, color='green')
plt.title('Distribution of Petal Length')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris_df)
plt.title('Petal Length vs. Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
# Pair plot to visualize relationships between all features
sns.pairplot(iris_df, hue='species')
plt.suptitle('Pairwise Relationships in the Iris Dataset', y=1.02)
plt.show()
```

```python
# Box plot of sepal width by species
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='sepal_width', data=iris_df)
plt.title('Sepal Width by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
# Violin plot of petal length by species
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='petal_length', data=iris_df)
plt.title('Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```
